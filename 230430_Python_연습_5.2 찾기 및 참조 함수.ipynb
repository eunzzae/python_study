### 인덱스로 값 확인  
    * 인덱스를 원래의 값으로 변환   
        - pd.map(function, iterable)

#### 예제   
    * 딕셔너리 변수 mapping을 생성하여, 공장의 지역을 지정  
    * 제품 코드의 오른쪽 끝자리 추출해 map() 함수에 Key로 입력하면 mapping 변수에서 해당 Key에 연결된 Value 반환하여 [공장]열에 추가

```python
product=pd.read_excel(r"D:\works\source\chapter05\식자재 주문.xlsx",sheet_name="product")
mapping={"1":"청주","2":"대구","3":"광주"}
product["공장"]=product["제품코드"].str[4].map(mapping)
product.head()
```

### 원하는 값 찾기  
    * 특정 위치에 해당하는 값을 참조   
        - pd.set_index("열 이름", inplace=True)

#### 예제  
    * 제품코드를 인덱스로 설정하며 ㄴ제품코드가 동일한 행의 데이터를 서로 참조함   
    * order에 제품명이 없어도 동일한 제품코드를 가진 product의 제품명을 가져올 수 있음(index참조)  
    
    * product에서 [제품명], [단가] 열을 가져와 order 에 추가하고 주문금액을 계산하는 예제

```python
product=pd.read_excel(r"D:\works\source\chapter05\식자재 주문.xlsx", sheet_name="product")
order=pd.read_excel(r"D:\works\source\chapter05\식자재 주문.xlsx", sheet_name="order")
product.set_index("제품코드", inplace=True)
order.set_index("제품코드",inplace=True)
order["제품명"]=product["제품명"]
order["단가"]=product["단가"]
order["주문금액"]=order["주문수량"]*order["단가"]
order
```
