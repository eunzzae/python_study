### 조건 함수 사용  
    * 조건식의 결과에 따라 별도 작업 수행  
        - 조건을 판단 후 특정 위치의 값을 변경하려면 for문과 if 문 사용  
        - for 문에 enumerate()함수를 같이 사용하면 데이터 프레임의 자료를 인덱스와 함께 불러올 수 있음   
        - if~elif~else -> 조건에 따라 코드 실행
        - enumerate(변수) -> 변수 내 값과 인덱스 반환 

```python
fruit=pd.read_excel(r"D:\works\source\chapter05\과일 주문.xlsx", sheet_name="Sheet1")
fruit["비고"]=""
for idx, x in enumerate(fruit["금액"]):
    if x>=500000:
        fruit["비고"].loc[idx]="50만원 이상"
    elif x>=200000:
        fruit["비고"].loc[idx]="20만원 이상"
    else:
        fruit["비고"].loc[idx]="20만원 미만"
fruit.head(6)
```

### 날짜 및 시간 함수   
    * 날짜형 포맷 출력 시 사용하는 datetime 자료형  
    * 날짜 계산 시 사용하는 timedelta 자료형   
        - pd.to_datetime() -> 문자열 데이터를 날짜형 데이터로 변환
        - pd.dt.day -> datetime자료형에서 일자만 출력
        - pd.dt.days -> timedelta자료형에서 일자만 출력
        - pd.datetime.now() -> 오늘 날짜 출력 

```python
fruit=pd.read_excel(r"D:\works\source\chapter05\과일 주문.xlsx", sheet_name="Sheet1")
pd.to_datetime(fruit["판매일자"])
pd.to_datetime(fruit["유통기한"])
fruit["판매일"]=fruit["판매일자"].dt.day
fruit["오늘 날짜"]=pd.datetime.now()
fruit["유통기한 경과일수"]=(fruit["오늘 날짜"]-fruit["유통기한"]).dt.days
fruit.head()
```

```python
from datetime import datetime
time1=datetime(2019,10,1,15,30,1)
time2=datetime.now()
print((time2-time1).days, "일")
print((time2-time1).seconds, "초")
print((time2-time1).seconds/3600,"시간")
```
