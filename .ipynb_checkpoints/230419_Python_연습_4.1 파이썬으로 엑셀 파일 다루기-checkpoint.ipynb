## 데이터 프레임 생성  
- Series(), DataFrame() 함수로 자료형 제공  
- 딕셔너리 유형으로 data 생성 후 , 데이터 프레임 만들기

```python
import pandas as pd
data={"이름":["홍길동","이순신","강감찬","임꺽정","이성계"],
      "출생연도":[1980,1986,1990,1985,1988],
      "점수":[1.5,1.7,3.6,2.4,2.9]}
df=pd.DataFrame(data)
df
```

```python
import pandas as pd
```

```python
import pandas as pd
data={"이름" : ["홍길동","이순신","강감찬","임꺽정","이성계"],
      "출생연도" : [1980,1986,1990,1985,1988],
      "점수" : [1.5,1.7,3.6,2.4,2.9]}
df=pd.DataFrame(data)
df
```

- rename()함수로 데이터 프레임 열 이름 변경 

```python
df=df.rename({"출생연도":"출생"},axis="columns")
df
```

## 행과 열 추가 및 삭제   

 ### 행과 열 데이터 가져오기

- 2개 이상 열 선택, df[["열 이름","열 이름"]]

- loc[], iloc[] 함수로 행과 열 데이터 가져오기 

```python
df.loc[0:1,["이름","출생"]]
```

```python
df.iloc[0:2,0:2]
```

### 열 추가 및 삭제

```python
df
```

- 열 추가는 열이름을 지정하고 리스트 형태로 데이터 직접 입력 또는 연산을 통해 새로운 열 생성 

```python
df["보너스"]=df["점수"]*5
df
```

```python
df["지역"]=["서울","서울","부산","대구","인천"]
df
```

```python
del df["보너스"]
df
```

- 행을 추가할 때는 loc[ ] 함수 사용

```python
df.loc[5]=["김순신",1980,3.3,"광주"]
df
```

- 특정행과 열에 있는 데이터 값만 변경 가능

- iloc[ ] 를 사용해 4행 1열의 데이터 변경, 새로 추가한 5행 삭제 시 drop()함수 사용

```python
df.iloc[4,1]=1999
df.drop(5,inplace=True)
df
```

- index()함수로 특정 조건을 만족하는 인덱스 값 구함

```python
df1=df.copy()
id=df1[df1["점수"]<=2.0].index
df1.drop(id,inplace=True)
df
```
