# 6.1 matplotlib으로 그래프 그리기  
 
 ### 그래프 종류   
     * 그래프 제목이나 범례 한글 사용하려면, 폰트 저장 필요
     - import matplotlib.puplot as plt
     plt,rcParams["font.family"]="Malgun Gothic"

```python
import matplotlib.pyplot as plt
plt.rcParams["font.family"]="Malgun Gothic"
```

### 선 그래프   
    * x, y축 데이터(변수)의 관계를 선으로 연결함   
    * plot() 함수로 선 그래프 그릴 수 있고, 컬러 및 라인 두께 지정 가능 
    - plt(x,y,color,lw)  # color=컬러, lw=라인두께  
    
    cf) -plt.plot([1,10],[1,1],linestyle="solid") - 실선
        -plt.plot([1,10],[2,2],linestyle="dashed") - 파선
        -plt.plot([1,10],[3,3],linestyle="dashdot") - 쇄선
        -plt.plot([1,10],[4,4],linestyle="dotted") - 점선

```python
plt.plot([1,10],[1,1],linestyle="solid")
plt.plot([1,10],[2,2],linestyle="dashed") 
plt.plot([1,10],[3,3],linestyle="dashdot") 
plt.plot([1,10],[4,4],linestyle="dotted") 
```

```python
height=[155,160,163,167,170,174,178,182,186,190]
weight=[44,46,48,50,57,62,70,74,79,82]
plt.title("선 그래프")
plt.xlabel("몸무게")
plt.ylabel("키")
plt.plot(weight,height, color="red", lw=3)
plt.show()
```

#### cf) 산점도 그리기  
    * plt.scatter()안에 marker="^"
    * 사각형(s), 삼각형(^), 엑스(x), 원(o), 별표(*)

```python
height=[155,160,163,167,170,174,178,182,186,190]
weight=[44,46,48,50,57,62,70,74,79,82]
plt.title("스캐터 플랏")
plt.xlabel("몸무게")
plt.ylabel("키")
plt.scatter(weight, height,marker="^")
plt.show()
```

```python
height=[155,160,163,167,170,174,178,182,186,190]
weight=[44,46,48,50,57,62,70,74,79,82]
plt.title("스캐터 플랏")
plt.xlabel("몸무게")
plt.ylabel("키")
plt.scatter(weight, height,marker="*")
plt.show()
```

### 막대 그래프  
    * plt.bar(x,y,width, color) width=그래프폭, color=색상  

```python
x=["사과","포도","딸기"]
y=[12,31,24]
plt.title("과일 생산량")
plt.bar(x,y,color="lightblue", width=0.5)
plt.xlabel("과일 종류")
plt.ylabel("판매량")
plt.ylim(0,40)
plt.show()
```

### 원 그래프  
    * pie()함수에서 비율을 나타내는 autopct속성을 "%1.1f%%"로 설정하면 항목별 비율을 소수점 첫째자리까지 표시  
    - plt.pie(y,labels, autopct,colors, shadow)  labels=항목, autopct=파이 조각 백분율, colors=파이조각 컬러, shadow=그림자 효과

```python
x=["사과","포도","딸기","참외"]
y=[12,31,24,46]
colors=["coral","cornsilk","pink","aqua"]
plt.title("원 그래프")
plt.pie(y,labels=x,autopct="%1.1f%%",colors=colors,shadow=True)
plt.show()
```

    * explode속성을 입력하면 입력한 크기만큼 분리됨

```python
x=["사과","포도","딸기","참외"]
y=[12,31,24,46]
colors=["coral","cornsilk","pink","aqua"]
plt.title("원 그래프")
plt.pie(y,labels=x,autopct="%1.1f%%",colors=colors,explode=(0.1,0,0,0), shadow=True)
plt.show()
```

### 히스토그램  
    * hist()함수로 히스토그램 그리고, x에 히스토그램 데이터를 입력하고 bins에는 계급구간수 입력   
    - plt.hist(x,bins,color)

```python
x=[18,4,10,22,19,-10,10,-2,-1,4,1,15,8,1,4,3,15,-2,3,-9,-26,7,9,-7,23,-15,0,-2,15,15]
plt.title("Histogram")
plt.grid(True)
plt.hist(x,bins=10,color="lightgreen")
plt.show()
```

### 상자 수염 그래프
    * 연속형 변수에 대해 분포 형태, 퍼짐 정도, 이상치 여부 등을 시각화하여, 하나의 그룹 혹은 여러 개 그룹을 비교하는데 유용함   
     - plt.boxplot(x,sym,vert)  sym=이상치 표시, vert=수직/수평 박스 플랏  

```python
x=[55,60,63,67,70,74,78,66,64,73,106]
plt.boxplot(x,sym="bo",vert=1)
plt.show()
```

```python
height=[155,160,163,167,170,174,178,182,186,190]
weight=[44,46,48,50,57,62,70,74,79,82]
plt.title("선 그래프")
plt.xlabel("몸무게")
plt.ylabel("키")
plt.plot(weight, height,color="blue",lw=1)
plt.scatter(weight,height, color="red",s=100)
plt.show()
```
