# 5.1 동적 배열 함수  

### 원하는 데이터 필터링 하기  

    * 데이터 프레임에서 조건에 맞는 행만 추출  
        - loc[], loc.isin() 함수를 이용  
            - loc[] 열 이름과 필터값을 조건식으로 비교하여 원하는 행 추출  
            - loc.isin() loc[]에 열이름을 지정하고 isin() 함수에 필터값 지정하여 원하는 행 추출  

        - pd.loc[pd["열"]=="필터값"]  
        - pd.loc[pd["열"].isin(["필터값"])

```python
import pandas as pd  
work = pd.read_excel(r"D:\works\source\chapter05\근무 유형.xlsx", sheet_name="Sheet1")
work
```

#### 예제  
    * loc[] 함수를 이용해, 부서명이 제조팀인 행을 추출하여 w1에 저장, loc.isin()함수를 이용해 근무형태가 상근인 행을 w2에 저장

```python
work=pd.read_excel(r"D:\works\source\chapter05\근무 유형.xlsx", sheet_name="Sheet1")
w1=work.loc[work["부서"]=="제조팀"]
w2=work.loc[work["근무형태"].isin(["상근"])]
display(w1,w2)
```

####  연산자 사용   
    * and(&)  
    * or(|)  
    * not(~) 

```python
work.loc[(work["부서"]=="제조팀")|(work["근무형태"]=="교대")]
```

```python
work.loc[~work["근무형태"].isin(["상근"])]
```

### 기준 열로 정렬하기   
    * 특정 열 기준으로 정렬 하기   
        - pd.sort_values(by="열", ascending=True/False)  
        - pd.sort_index(ascending=True/False)

#### 예제   
    - 사원번호를 기준으로 오름차순 정렬, 인덱스 기준으로 정렬 

```python
work=pd.read_excel(r"D:\works\source\chapter05\근무 유형.xlsx",sheet_name="Sheet1")
work.sort_values(by="사원번호").head()
```

```python
work=pd.read_excel(r"D:\works\source\chapter05\근무 유형.xlsx",sheet_name="Sheet1")
work.sort_index().head()
```

### 중복 행 제거  
    * 행 전체가 중복되거나 일부 값이 중복되는 경우   
        - pd.duplicated() - 중복되는 행이 있을 경우 True 출력  
        - pd.drop_duplicates() - 중복되는 행 제거   

```python
work=pd.read_excel(r"D:\works\source\chapter05\근무 유형.xlsx", sheet_name="Sheet1")
work.duplicated()
```

```python
work.drop_duplicates()
```
