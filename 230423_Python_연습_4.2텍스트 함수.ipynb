### 여러 셀의 문자 합치기  

* 엑셀 : concatenate()함수를 이용하면 됨  
* 파이썬 : 두가지 방법이 있음  (+와 sum을 이용)
  - pd["열3"] = pd["열1"]+pd["열2"]  
  - pd["열3"] = pd[["열1","열2"]].sum(axis=0 or 1)  
    axis=0이면 행에 추가, axis=1이면 열에 추가 

```python
info=pd.read_excel(r"D:\works\source\chapter04\직원 정보.xlsx",sheet_name="Sheet1")
info["성명"] = info["성"] + info["이름"]
info["성명1"] = info[["성","이름"]].sum(1)
info
```

### 몇 개의 문자만 추출하기  
 * 특정 위치나 범위에 있는 문자열을 추출  
 * 엑셀 : left(), right()함수사용  
 * 파이썬 : 추출할 특정 문자열의 위치나 범위를 []안에 지정  
   - str[]로 위치나 범위지정  ex. pd.str[0:6] --> 0부터 5번째까지의 문자열 추출  
    - pd["열"].str[위치]  
    - pd["열"].str[시작위치:끝위치]

```python
info=pd.read_excel(r"D:\works\source\chapter04\직원 정보.xlsx",sheet_name="Sheet1")
info["사원번호 앞 4자리"]=info["사원번호"].str[0:4]
info["전화번호 뒤 4자리"]=info["전화번호"].str[9:13]
info
```

#### 참고  
 - 주소가 공백으로 구분되어 있을 때, split()함수 사용하여 문자열만 추출  
   * 주소 열을 split(" ")을 통해 공백 분리하여 str[1]로 첫번째 문자열 추출

```python
info["구"]=info["주소"].str.split(" ").str[1]
info[["성","이름","전화번호","구"]]
```

### 영문 대소문자 바꾸기  
 - 영문자 입력된 셀 전체를 대소문자로 변환, 문자열의 첫글자만 대문자 변환  
  * upper(), lower(), capitalize()  
  * pd["열"].str.upper(), pd["열"].str.lower()  
    pd["열"].str.capitalize()  
   
    <참고>  
     * str.swapcase() 대문자는 소문자로 소문자는 대문자로 변환

```python
info=pd.read_excel(r"D:\works\source\chapter04\직원 정보.xlsx",sheet_name="Sheet1")
info["대문자"]=info["영문명"].str.upper()
info["첫 글자"]=info["영문명"].str.capitalize()
info[["순번","성","이름","영문명","전화번호","대문자","첫 글자"]]
```

```python
str.swapcase("ABcde")
```

### 특정 문자 바꾸기  
- 다른 문자로 변경할 때  
 * str.replace() 찾은 문자를 새로운 문자로 대체하여, 값이 중복일 경우 변경횟수까지 지정 (변경횟수를 생략할   시 찾은 문자 모두 대체)   
- pd["열"].str.replace(찾을 문자, 변경할 문자, 변경 횟수)

```python
info=pd.read_excel(r"D:\works\source\chapter04\직원 정보.xlsx",sheet_name="Sheet1")
info["phone"]=info["전화번호"].str.replace("-","",1)
info["phone1"]=info["전화번호"].str.replace("-"," ",)
info[["순번","성","이름","전화번호","phone","phone1"]]
```

### 문자열 길이 구하기  
 - 문자열의 길이 구하기  
 - 문자열에 포함된 공백도 문자열 길이에 포함
  * pd["열"].str.len()   
  
  <참고>
   * head(행 숫자), tail(행 숫자) 원하는 행만큼 결과 출력(숫자 생략 시 각 5행씩 결과 출력)

```python
info=pd.read_excel(r"D:\works\source\chapter04\직원 정보.xlsx",sheet_name="Sheet1")
info["주소 길이"]=info["주소"].str.len()
info[["순번","성","이름","주소","주소 길이"]].head()
```

```python
info=pd.read_excel(r"D:\works\source\chapter04\직원 정보.xlsx",sheet_name="Sheet1")
info["주소 길이"]=info["주소"].str.len()
info["공백 제거"]=info["주소"].str.strip()
info["공백 제거 후 길이"]=info["공백 제거"].str.len()
info[["순번","주소","주소 길이","공백 제거","공백 제거 후 길이"]].tail(6)
```
