### 데이터 합계 구하기   
* sum()의 함수로 데이터 프레임의 열, 행 방향 합계 구하기  

 - axis=0은 행방향으로 각 열을 더해 범주형, 수치형 데이터 모든 합계 산출  --> 열을 더함
 - axis=1은 열 방향으로 각 행의 합계 더함 --> 행을 더함   
 
    * pd["열"].sum(axis=0 or 1)  --> 행 방향 열(0) 또는 열 방향 행(1) 합계 구하기   
    * pd["열"]+pd["열"]+...+pd["열"]  --> +연산자로 직접 계산

열의 합계를 구하는 식이며, axis=0으로 지정   
 * axis의 디폴트 값은 0이므로 생략가능  

```python
score=pd.read_excel(r"D:\works\source\chapter04\성적 처리.xlsx",sheet_name="Sheet1")
total_korea=score["국어"].sum(0)
total_korea
```

  * score.iloc[:,2:7].sum(1)은 score 데이터 프레임 변수의 모든 행과 2~6열을 선택하여, 행의 데이터 합계를 구함

```python
score["sum"]=score.iloc[:,2:7].sum(1)
score["sum1"]=score["국어"]+score["영어"]+score["수학"]+score["사회"]+score["과학"]
score.head()
```

### 데이터 평균 구하기  

 - 데이터 평균은 통계분석 시 가장 많이 활용하며, mean()함수를 이용

- 데이터 프레임의 행과 열 방향 평균은 mean()을 이용   
 * axis=0으로 설정 시 열을 구하고, axis=1 설정 시 행을 구함
 
   * pd["열"].mean(axis=0 or 1)  
   * (pd["열"]+pd["열"]+...+pd["열"])/열 개수

```python
score=pd.read_excel(r"D:\works\source\chapter04\성적 처리.xlsx",sheet_name="Sheet1")
korea_avg=score["국어"].mean()
korea_avg
```

```python
score["평균"]=score.iloc[:,2:7].mean(1)
score["평균1"]=(score["국어"]+score["영어"]+score["수학"]+score["사회"]+score["과학"])/5
score
```

### 조건에 따른 합계, 평균 구하기  

- 조건에 따라 항목별 합계 및 평균 구하기   
 
 * pd.groupby(["조건 열"]).sum() --> 합계     
 * pd.groupby(["조건 열"]).mean() --> 평균

```python
score=pd.read_excel(r"D:\works\source\chapter04\성적 처리.xlsx",sheet_name="Sheet1")
score1=score.groupby(["반"]).sum()
score1
```

```python
score2=score.groupby(["반"]).mean()
score2
```

```python
score=pd.read_excel(r"D:\works\source\chapter04\성적 처리.xlsx", sheet_name="Sheet1")
score["평균"]=score.iloc[:,2:7].mean(1)
score["순위_오름"]=score["평균"].rank(ascending=True)
score["순위_내림"]=score["평균"].rank(ascending=False)
score["순위_내림_min"]=score["평균"].rank(method="min",ascending=False)
score.head(12)
```
